```python
scan_rates = [1000, 500, 200, 100, 50, 5] 
```

```python
# Polished
plt.figure(figsize=(8, 6))
cmap = plt.colormaps['tab10']

datasets1 = [
    Po_1000mV_4,
    Po_500mV_4,
    Po_200mV_4,
    Po_100mV_4,
    Po_50mV_4,
    Po_5mV_4
]

for i, (df, rate) in enumerate(zip(datasets1, scan_rates)):
    plt.scatter(
        df['E'],
        df['I'] * 1e6,
        s=1,
        color=cmap(i),
        label=f"{rate} mV/s"
    )

plt.xlabel("E (V vs Ag/AgCl (1 M KCl))", fontsize=22, fontweight='bold')
plt.ylabel("I (μA)", fontsize=22, fontweight='bold') 

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_color('black')
    spine.set_linewidth(2)

ax.tick_params(axis='both', which='major', labelsize=14, direction='out', width=2, length=8)
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontweight('bold')

plt.legend(
    scatterpoints=1, markerscale=5, fontsize=14, title_fontsize=14, frameon=False,
    loc='upper left'
)

plt.tight_layout()
plt.show()
```
